```python
## 不想用垃圾pandas
import json
with open("chinese-poetry/quan_tang_shi/json/001.json", encoding='utf-8') as f:
    a = json.load(f)
    print(a)
```

```python
poetries = []
for i in range(1, files_cnt+1):
    t_p = path_str+ f"{i:0>3}.json"
    # print(t_p)
    t_df = json.load(open(t_p))
    poetries+=t_df
print(poetries[0:2])
len(poetries)
```

```python
# 做字典
import collections

中文词频 = dict()
# 中文字典 = collections.OrderedDict() # 红黑树哈哈
for 诗歌 in poetries:
    for 中文 in 诗歌.values():
        if type(中文)==type(""):
            for 汉字 in 中文:
                中文词频[汉字] = 中文词频.get(汉字, 0) + 1
        elif type(中文)==type(['']):
            for paragraph in 中文:
                for 汉字 in paragraph:
                    中文词频[汉字] = 中文词频.get(汉字, 0) + 1
print(中文词频['月'])
len(中文词频)
```

```python
print(中文词频)
中文词频['橈']
```

```python
items = list(中文词频.items())
items.sort(key=lambda x:x[1], reverse=True)
print(items[-5:-1])
items[0:5]

```

```python
中文到整数转换表 = dict()
for i, item in enumerate(items):
    中文到整数转换表[item[0]] = i
中文到整数转换表['月']
```

```python
with open('中文到整数转换表.json', 'w') as f:
    json.dump(中文到整数转换表, f)
```

```python
## 生成作者表
作者频率 = {}
for 诗歌 in poetries:
    作者 = 诗歌['author']
    作者频率[作者] = 作者频率.get(作者,0)+1
print(作者频率['李白'])
print(len(作者频率))
print()
bad_author = ['不詳', '佚名', '無名氏', '無名', '匿名'
    , '子虚乌有', '張三', '子虛烏有', '某某', '某君', '逸名'
    , '不详', '无名氏']
for i in bad_author:
    print(作者频率.get(i, 0))
    if i in 作者频率:
        作者频率.pop(i)
# 去除掉不详和无名氏！
print()
for i in bad_author:
    print(作者频率.get(i, 0))
```

```python
items = list(作者频率.items())
items.sort(key=lambda x:x[1], reverse=True)
print(items[-100:-1])
items[0:5] # 齐己=齊己
```

```python
作者到整数转换表 = dict()
for i, item in enumerate(items):
    作者到整数转换表[item[0]] = i
作者到整数转换表['李白']
```

```python
with open('作者到整数转换表.json', 'w') as f:
    json.dump(作者到整数转换表, f)
```

```python
有传记的 = []
有注解的 = []
for 诗歌 in poetries:
    if 诗歌['author'] in bad_author:
        continue
    if 诗歌['biography']!="":
        有传记的.append(诗歌)
    elif 诗歌['notes']!=[""]:
        有注解的.append(诗歌)
print(len(有传记的))
print(len(有注解的))
有注解的[-5:-1] # 好像没啥用
```

```python
数字化全唐诗 = []
for 诗歌 in poetries:
    作者 = 诗歌['author']
    if 作者 in bad_author: continue

    数字化诗歌 = dict()
    数字化诗歌['title'] = [中文到整数转换表[c] for c in 诗歌['title']]
    数字化诗歌['author'] = 作者到整数转换表[作者]
    # 简化 paragraphs ， 已经有句号了，就不用那个了。
    content = "".join(诗歌['paragraphs'])
    数字化诗歌['content'] = [中文到整数转换表[c] for c in content]
    数字化全唐诗.append(数字化诗歌)
数字化全唐诗[0:5]
```

```python
with open('数字化全唐诗.json', 'w') as f:
    json.dump(数字化全唐诗, f)
```
